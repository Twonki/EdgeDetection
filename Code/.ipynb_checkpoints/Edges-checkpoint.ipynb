## Simple Edge Filter

After we've looked at our picture and it's size, we can already start using our simple filter on the image and see what it says. For simplicity I put both plots for x and y in one line.

The function "convolve1d" applies a 1-dimensional array as convolutionkernel. The parameter axis determines which axis (y or x) it's supposed to iterate. Therefore declaring our simple filter once is enough for both operations.
The standard-padding type is *same*, telling our convolution that the paddings of the image are filled with the nearest neighboor. 

```python
simpleFilter = [-1,0,1] # We leave the coefficient of 0.5 for now, we put it in later

xEdges = ndimage.filters.convolve1d(img,simpleFilter)
yEdges = ndimage.filters.convolve1d(img,simpleFilter,axis=0);

figure,(aX,aY) = plt.subplots(1,2,sharey=True)
aX.set_title('X-Edges')
aX.imshow(xEdges)
aY.set_title('Y-Edges')
aY.imshow(yEdges)
plt.show()
```

We can see that there are some differences in the image *ascent*. 
If we work with Pikachu, we see stronger differences e.g. the mouth of pikachu. 

Let's define required methods. The formulas are shown in the presentation. 

Some of the operations, e.q. the np.sqrt and np.power keep the dimensions of the array (therefore the result is still a visible image). 
Working on two ndimages requires the nditerator, which sadly only returns an list. We need to reparse the list to an array, and reassign the shape (of the image). This procedure shows up more often throughout the notebook.

The *simpleEdgeDetection* has the coefficients this time. The output is the magnitude of the images shown above.

```python
def magnitude(xEdges,yEdges):
    return np.sqrt(np.power(yEdges,2) + np.power(xEdges,2))

def direction(xEdges,yEdges):
    return np.asarray([np.max(np.arctan([a,b])) for a,b in np.nditer([xEdges,yEdges])]).reshape(xEdges.shape)

def simpleEdgeDetection(img):
    simpleFilter = [-1,0,1]
    xEdges = ndimage.filters.convolve1d(img,simpleFilter) * 0.5
    yEdges = ndimage.filters.convolve1d(img,simpleFilter,axis=0) * 0.5
    # the Result must be casted using 'dtype=uint8', as floats cannot be displayed in black and white image.
    return np.asarray(magnitude(xEdges,yEdges),dtype="uint8")

plt.imshow(simpleEdgeDetection(img))
```

#  Advanced Edge Detection
Ok, after we're done with our 1D Kernels, we can step on further to 2D Kernels. The next example shows the output of a prewitt filter.

We now use the Method *convolve*, which doesn't require an axis. The padding behaviour is the same as for convolve1D.

Using *ascent*, you can see big differences for example in the upper part of the image.

```python
prewittX = ndimage.filters.convolve(img,kernels.prewittX)/6 #Prewitts Coeff is 1/6
prewittY = ndimage.filters.convolve(img,kernels.prewittY)/6
prewitt = magnitude(prewittX,prewittY)

prewitt = np.asarray(prewitt,dtype='uint8') # Floats cannot be plotted as gray images :(
plt.imshow(prewitt)
plt.show()
```

Let's simplify this behaviour and put it in a method. 

We then check it on the *sobel* operator. 

```python
def edgeDetection(img,xKernel,yKernel,coeff):
    xEdges = ndimage.filters.convolve(img,xKernel)*coeff
    yEdges = ndimage.filters.convolve(img,yKernel)*coeff
    mag = magnitude(xEdges,yEdges)
    dirs = direction(xEdges,yEdges)
    return mag,dirs
```

```python
sobel,sobeldirs = edgeDetection(img,kernels.sobelX,kernels.sobelY,0.125)
figure,(m,d) = plt.subplots(1,2,sharey=True)
m.set_title('Magnitude')
m.imshow(sobel)
d.set_title('Direction')
d.imshow(sobeldirs)
plt.show()
```

Note: the direction is mapped on a normalized basis. 

The values \[0,PI\] are mapped to \[0,255\] with a simple linear function. 

## Compass Operators

Ok, lets go to the next part: Compass Operators. 

Remember: we apply 8 kernels for different directions on the *compass*. For the magnitude we pick the maximum of the found values, and we denote in a seperate array which kernel found this maximum. 

The *degree* of the edge is directly related to the kernel which found the maximum. If the "X-Right"-Kernel found the maximum, the degree is 0. If the "Diagonal Rising"-Kernel found the maximum, the degree is 0.45 or PI/4

**Compass-function**

The following function is a little more complex, so I describe it seperatly. 

The function requires an image, an array of the filter-kernels (which need to be in an ordered fashion!) and the coefficient of the kernel. 

First, we produce an array of images for the edge kernels. We recieve an (8,img.X,img.Y) shaped array. Each contains 1 pixel of the magnitude of the related edgedirection. 

Second, and this is the very nicest I've ever seen in python, we transpose the (8,img.X,img.Y)-matrix to a (img.X,img.Y,8)-matrix. So to say, we've got ourselves a *super-Image*, with a list of pixels for the positions. So if we access the element (1,5) we get and array of length 8 containing the pixel-values of every single image at the point (1,5).  

These superpixels are still ordered the way the filter-kernel was ordered.
We store this result in the variable *helper*. 

From then we create two images: We pick for the maximum the max of our superpixels, reducing the dimension to (img.X,imgY). 
We also pick the *argmax* of the superpixel, to save which kernel found the maximum. 

```python
def compass(img, filterarray, coeff):
    imgArr = np.asarray([ndimage.filters.convolve(img,filter)*coeff for filter in filterarray])
    helper = imgArr.transpose(1,2,0)
    magnitude = helper.max(axis=2)
    compass = helper.argmax(axis=2)
    return magnitude,compass
```

Let's bring this glorious function to life using the extended sobel compass-operator. You can look it up in the "kernels.py". 

```python
mag,com = compass(img,np.asarray(kernels.extendedSobel),0.125)
figure,(m,c) = plt.subplots(1,2,sharey=True)
m.set_title('Magnitude')
m.imshow(mag)
c.set_title('Compass')
c.imshow(com)
plt.show()
```

*Note*: the *compass* image is mapped from \[0,7\] to \[0,255\]. To improve this output, you should consider implementing/using a function that maps these 8 values to different, and easy to differ, RGB-Values. 

To close this chapter, we take a look at the kirsch-operator.

```python
mag,com = compass(img,np.asarray(kernels.kirsch),0.0333)
figure,(m,c) = plt.subplots(1,2,sharey=True)
m.set_title('Magnitude')
m.imshow(mag)
c.set_title('Compass')
c.imshow(com)
plt.show()
```

## Canny-Edge
We now implement the canny-edge operator, following the recipe presented in the talk:

* Preprocess using gaussian blur and simple edge-detection
* Edge-Localization looking for maximums
* hysteresis threshholding

we start with the preprocessing. 

We apply the gaussian-filter using the build in function from scipy and then use our own method declared earlier. Let's take a look. 

```python
def preprocessCannyEdge(img,sigma=2):
    gaussed = ndimage.filters.gaussian_filter(img,sigma)
    return simpleEdgeDetection(gaussed)

figure,(s1,s2,s3) = plt.subplots(1,3,sharey=True)
s1.set_title('Sigma=1')
s1.imshow(preprocessCannyEdge(img,sigma=1))
s2.set_title('Sigma=2')
s2.imshow(preprocessCannyEdge(img,sigma=2))
s3.set_title('Sigma=3')
s3.imshow(preprocessCannyEdge(img,sigma=3))
plt.show()
```

Preprocession looks fine. The next part is the **edge-localization**. 

Pixels are choosen if they are a local maximum in (atleast) one direction, or are set to 0 if not. 

We split this behaviour in two functions: one helps us finding the maxima in our edge-image, and one actually applying it to the image.

**MaxMask-Function**:
    
The maxmask function takes an image and returns an same-shaped boolean array, whether the pixel is a local maximum or is not. 

We start using the same procedure as in the compass operators, this time we use a filterarray containing our simple-edge filters for different directions. If any compass results in zero, we have a local extrema. 

We also need to look at the second derivation: We apply the laplacian-filter on our image, and double check if the value for the found extrema is positive. if the value is in fact positive, we have a local maximum. 

The function *edgelocalization* takes the bool-mask and the image, and returns only images which are marked as local maxima. 

```python
def maxMask(img):
    filters = np.asarray(kernels.maxCompass)
    imgArr = np.asarray([ndimage.filters.convolve(img,f) for f in filters])
    helper = imgArr.transpose(1,2,0)
    # we need a different shape than in the compass, to work properly with the logic 'in' operator
    helper= helper.reshape(helper.shape[0]*helper.shape[1],8)
    #Ravel unfolds to a simple list
    laplace = ndimage.filters.convolve(img,kernels.laplacian).ravel()
    
    mask = [(0 in tupel) for tupel in helper]
    mask = np.asarray([(t and l>=0) for (t,l) in np.nditer([mask,laplace])]).reshape(img.shape)
    return mask

def edgeLocalization(edgeImg,mask):
    return np.asarray([v if m else 0 for v,m in np.nditer([edgeImg,mask])]).reshape(edgeImg.shape)
```

Lets put these methods to action. 

*Note*: simple tests show that the results are much better for higher sigmas then shown here. you can play around it for yourselves, I just wanted to keep it parallel to the example given above

```python
edgeImg1 = preprocessCannyEdge(img,sigma=1)
mask1 = maxMask(edgeImg1)
edgeImg2 = preprocessCannyEdge(img,sigma=2)
mask2 = maxMask(edgeImg2)
edgeImg3 = preprocessCannyEdge(img,sigma=3)
mask3 = maxMask(edgeImg3)

figure,(s1,s2,s3) = plt.subplots(1,3,sharey=True)
s1.set_title('Sigma=1')
s1.imshow(edgeLocalization(edgeImg1,mask1))
s2.set_title('Sigma=2')
s2.imshow(edgeLocalization(edgeImg2,mask2))
s3.set_title('Sigma=3')
s3.imshow(edgeLocalization(edgeImg3,mask3))
plt.show()
```

The **hysteresis-threshhold** consists of two parts: 

1. find every pixel>t0 and put it in "ok"-list
2. find every neighboor to every pixel in the "ok" list, where neighboor>t1
3. repeat 2.

Step 1 is simple and the frist line of our code. We call the "ok"-List our *hysteresis mask* and set it initially to our pixels from step 1. 

We also need the variable *lastmask*. 
We need it to compare, if our last iteration changed anything to our hysteresis mask. We initalize it with zeros. 

We do so by using the XOR operator on both masks. If there are different values at position x, the XOR operator will return *True* for x. the operator *any* checks if the array contains any *True*. If there are no changes and every bit is exactly the same, the XOR  will return an array full of *False* and any will fail aswell. 

**the while loop**

In the loop, we first update the lastMask to our current hysteresis mask. 

We then find the possible neighboors of our "ok"-list by Dilation. The result of a binary-dilation is a same-shaped array, where every neighboor of a true value is set true. 

We then check the possible neighboors if they are greater than t1, and if so, update our hysteresis mask. We repeat the while loop. 

When our hysteresis mask stops changing, we apply the mask to the image and return the result. Applying the mask means returning the value if it's in our mask, or 0 if it's not. 

```python
def hysteresisThreshhold(edgeImg, t0, t1):
    evertrues = np.asarray([p>=t0 for p in edgeImg]).reshape(edgeImg.shape)
    hysteresisMask = evertrues.copy()
    lastMask = np.asarray(np.zeros(edgeImg.shape),dtype=bool)
    while any(np.bitwise_xor(lastMask.ravel(),hysteresisMask.ravel())): 
        lastMask = hysteresisMask
        possibleNeighboors = ndimage.binary_dilation(hysteresisMask)
        hysteresisMask = np.asarray([p>=t1 and n for (p,n) in np.nditer([edgeImg,possibleNeighboors])]).reshape(edgeImg.shape)
    masked = np.asarray([p if m else 0 for p,m in np.nditer([edgeImg,hysteresisMask])]).reshape(edgeImg.shape)
    return masked
```

We now can finally use our selfmade canny-edge-operator ! 

I've put the declaration and the use in different boxes, so we can stop the time properly

```python
def cannyEdge(img,t0,t1,sigma=2):
    edgeImg = preprocessCannyEdge(img,sigma)
    mask = maxMask(edgeImg)
    localizedEdges = edgeLocalization(edgeImg,mask)
    final = hysteresisThreshhold(localizedEdges,t0,t1)
    return final
```

```python
%%time
plt.imshow(cannyEdge(img,8,3,sigma=5))
```

## Edge Sharpening
to finish this notebook, we look into the two introduced methods to sharpen edges. 

* Sharpening with second derivation (Laplace)
* Unsharp masking using Gaussian Blur

To apply **laplacian**-sharpening we need to calculate the 2nd derivation, weight it with a factor w, and then apply it to our ur-image. 

We put it in a simple method

```python
def laplacianSharpening(img,w):
    laplacian=ndimage.filters.convolve(img,kernels.laplacian)
    laplacianWeighted= laplacian*w
    return img+laplacianWeighted

figure,(s1,s2,s3) = plt.subplots(1,3,sharey=True)
s1.set_title('w=0.1')
s1.imshow(laplacianSharpening(img,0.1))
s2.set_title('w=0.25')
s2.imshow(laplacianSharpening(img,0.25))
s3.set_title('w=0.5')
s3.imshow(laplacianSharpening(img,0.5))
plt.show()
```

For the **unsharp masking** we simply follow our recipe:

1. smooth the image
2. subtract the smoothed image from the raw image, call it mask
3. add the mask with a weight w to the raw image

```python
def unsharpMasking_Gauss(img, w, sigma=3):
    smoothed = ndimage.filters.gaussian_filter(img,sigma)
    mask = img - smoothed
    return img+mask*w
```

```python
figure,axArr = plt.subplots(3,3,sharey=True,sharex=True)
axArr[0,0].set_title('w=0.1, sigma=1')
axArr[0,0].imshow(unsharpMasking_Gauss(img,0.1,sigma=1))
axArr[0,1].set_title('w=0.25, sigma=1')
axArr[0,1].imshow(unsharpMasking_Gauss(img,0.25,sigma=1))
axArr[0,2].set_title('w=0.5, sigma=1')
axArr[0,2].imshow(unsharpMasking_Gauss(img,0.5,sigma=1))
axArr[1,0].set_title('w=0.1, sigma=2')
axArr[1,0].imshow(unsharpMasking_Gauss(img,0.1,sigma=2))
axArr[1,1].set_title('w=0.25, sigma=2')
axArr[1,1].imshow(unsharpMasking_Gauss(img,0.25,sigma=2))
axArr[1,2].set_title('w=0.5, sigma=2')
axArr[1,2].imshow(unsharpMasking_Gauss(img,0.5,sigma=2))
axArr[2,0].set_title('w=0.1, sigma=3')
axArr[2,0].imshow(unsharpMasking_Gauss(img,0.1,sigma=3))
axArr[2,1].set_title('w=0.25, sigma=3')
axArr[2,1].imshow(unsharpMasking_Gauss(img,0.25,sigma=3))
axArr[2,2].set_title('w=0.5, sigma=3')
axArr[2,2].imshow(unsharpMasking_Gauss(img,0.5,sigma=3))
plt.show()
```
